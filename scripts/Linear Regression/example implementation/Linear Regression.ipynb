```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
x= DataFrame(data, columns = ['production_budget_usd'])
y= DataFrame(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=.4)
plt.xlim(0,4.5e8)
plt.ylim(0,3e9)
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.title("Cost v/s Revenue")
plt.show()
```

```python
# y = theta_0 + theta_1*X
rgr = LinearRegression()
rgr.fit(x,y)
```

```python
#theta_1 or slope
th0= rgr.coef_ 
th0
```

```python
#theta_0 or c
th1 = rgr.intercept_
th1
Y=rgr.predict(x)
x
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=.3)

plt.scatter(x,Y) 
plt.plot(x['production_budget_usd'],Y,color='black')  #to convert dataframe to an array 
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.title("Cost v/s Revenue")

plt.show()
```
